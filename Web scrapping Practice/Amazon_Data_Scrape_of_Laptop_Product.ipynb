```python
products = html_data('div',{'data-component-type':'s-search-result'})
```

```python
for product in products:
    #print(product)
    title = product.find('span',{'class':'a-size-medium a-color-base a-text-normal'}).text
    print(title)
    img = product.find('img')['src']
    print(img)
    rating = product.find('span',{'a-icon-alt'})
    # Ratting have nun value then it make problem
    if rating == None:
        rating = 'No Rating'
    else:
        rating = product.find('span',{'a-icon-alt'}).text
    print(rating)
    price = '' + product.find('span',{'class':'a-price-whole'}).text
    print(price)
    
        
    print('=======================================================================')
    print('=======================================================================')
```

```python
titles = []
imgs = []
ratings = []
prices = []

# Instead of print we store the values in list

for product in products:
    #print(product)
    title = product.find('span',{'class':'a-size-medium a-color-base a-text-normal'}).text
    titles.append(title)
    img = product.find('img')['src']
    imgs.append(img)
    rating = product.find('span',{'a-icon-alt'})
    # Ratting have nun value then it make problem
    if rating == None:
        rating = 'No Rating'
    else:
        rating = product.find('span',{'a-icon-alt'}).text
    ratings.append(rating)
    price = '' + product.find('span',{'class':'a-price-whole'}).text
    prices.append(price)
    
```

```python
import time
from tqdm import tqdm ### tqdm gives the progressbar
```

```python
titles = []
imgs = []
ratings = []
prices = []

# Instead of print we store the values in list
for i in tqdm(range(No_of_pages)):
    
        url = 'https://www.amazon.in/s?k=laptops&crid=9XOMCI4DFXN3&sprefix=laptops%2Caps%2C241&ref=nb_sb_noss_1&page='+str(i+1)
        driver.get(url)
        time.sleep(3)
        html_data = BeautifulSoup(driver.page_source, 'html.parser')
        products = html_data('div',{'data-component-type':'s-search-result'})
        #print(product)


    
        for product in products:
            title = product.find('span',{'class':'a-size-medium a-color-base a-text-normal'}).text
            titles.append(title)
            img = product.find('img')['src']
            imgs.append(img)
            rating = product.find('span',{'a-icon-alt'})
            # Ratting have nun value then it make problem
            if rating == None:
                rating = 'No Rating'
            else:
                rating = product.find('span',{'a-icon-alt'}).text
            ratings.append(rating)
            price = '' + product.find('span',{'class':'a-price-whole'}).text
            prices.append(price)
        
```

```python
len(titles)
```

### Store the data using pandas library

```python
Amazon_laptop = pd.DataFrame({'titles':titles, 'images':imgs, 'ratings':rating, 'prices':prices})
Amazon_laptop
```
