# if i want to find in between probability # p(697<=x<=740)

```python
mean = 711
std = 29

a = stats.norm.cdf(697, loc = mean, scale = std)
b = stats.norm.cdf(740, loc = mean, scale = std)

print('prob that the value is les than or equal to 697', round(a*100,2))
print('prob that the value is les than or equal to 740', round(b*100,2))
print("ANS",round((b-a)*100,2))
```
