### Cross Validation 

```python

train_acc = []
test_acc = []

for i in range(1,101,1):
    x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=i)
    knn.fit(x_train,y_train)
    y_pred_train = knn.predict(x_train)
    y_pred_test = knn.predict(x_test)
    train_acc.append(np.sqrt(accuracy_score(y_train, y_pred_train).round(2)))
    test_acc.append(np.sqrt(accuracy_score(y_test, y_pred_test).round(2)))
```

```python
print("Cross Validation training accuracy:", np.mean(train_acc).round(2))
print("Cross Validation testing accuracy:", np.mean(test_acc).round(2))
```

### For loop for diffrent k value

```python
avg_train_acc = []
avg_test_acc = []

train_acc = []
test_acc = []

for k in range(3, 16):
  knn = KNeighborsClassifier(n_neighbors=k)

  for i in range(1, 101):
    X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.3, random_state=i)
    knn.fit(X_train, Y_train)
    Y_pred_train = knn.predict(X_train)
    Y_pred_test = knn.predict(X_test)
    train_acc.append(accuracy_score(Y_train, Y_pred_train))
    test_acc.append(accuracy_score(Y_test, Y_pred_test))

  print(f"k: {k}")
  print("Cross validation training accuracy:", np.mean(train_acc).round(2))
  avg_train_acc.append(np.mean(train_acc).round(2))
  print("Cross validation test accuracy:", np.mean(test_acc).round(2))
  avg_test_acc.append(np.mean(test_acc).round(2))
  print()
```

### Find Accuracy using diffrent distances methods  

```python
metrics = ['euclidean', 'manhattan', 'chebyshev', 'minkowski']

for metric in metrics:
    if metric == 'mahalanobis':
        # Mahalanobis distance requires the inverse of the covariance matrix of the data
        knn = KNeighborsClassifier(n_neighbors=5, metric=metric, metric_params={'V': np.cov(X_train)})
    else:
        knn = KNeighborsClassifier(n_neighbors=5, metric=metric)
    
    # Fit the classifier to the training data
    knn.fit(x_train, y_train)
    
    # Evaluate the classifier on the test data
    accuracy = knn.score(X_test, y_test)
    print(f'Accuracy using {metric} distance metric: {accuracy:.2f}')
```

### Gridsearch to find Best Model Parameter 

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, confusion_matrix


model = KNeighborsClassifier()
model.fit(x_train, y_train)
model.predict(x_train)
model.predict(x_test)
# Define parameter grid
param_grid = {
    'n_neighbors': [5,7,9,11,13,15],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

def custom_scorer(y_true, y_pred):
    tn, fp, fn, tp = confusion_matrix(y_true, y_pred).ravel()
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    return (sensitivity + specificity) / 2  # average of sensitivity and specificity

# Make scorer from the custom scoring function
scorer = make_scorer(custom_scorer)

# Perform grid search using cross-validation
grid_search = GridSearchCV(knn, param_grid, cv=5, scoring=scorer)
grid_search.fit(X_train, y_train)

# Print best parameters and corresponding score
print("Best parameters found:", grid_search.best_params_)
print("Best score:", grid_search.best_score_)

# Evaluate best model on test data
best_model = grid_search.best_estimator_
test_score = best_model.score(X_test, y_test)
train_score = best_model.score(X_train, y_train)
print("Test accuracy of best model:", test_score)
print("Train accuracy of best model:", train_score)
```
