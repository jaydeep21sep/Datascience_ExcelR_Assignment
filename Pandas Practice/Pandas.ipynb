### Creation of Series from Scalar Values

```python
# A Series can be created using scalar values as shown in the example below: By default index
import pandas as pd #import Pandas with alias pd
series1 = pd.Series([10,20,30]) #create a Series
print(series1) #Display the series
```

```python
# User Define index
series2 = pd.Series(["Kavi","Shyam","Ravi"], index=[1,2,3])
print(series2) #Display the series
```

```python
# User define index as strring
series2 = pd.Series([2,3,4],index=["Feb","Mar","Apr"])
print(series2) #Display the series
```
